```python
import json
import matplotlib.pyplot as plt
import pandas as pd
import re
```

```python
data=[]
with open("arxiv-metadata-oai-2019.json",'r') as f:
    for idx,line in enumerate(f):
        d=json.loads(line)
        d={'abstract':d['abstract'],'categories':d['categories'],'comments':d['comments']}
        data.append(d)
data=pd.DataFrame(data)
```

```python
data
```

```python
def readfile(path,columns={'id','submitter','authors','title','comments','journal-ref','doi','report-no','categories','license','aabstract,versions','update_date','authors-parsed'},count=None):
    data=[]
    with open(path,'r') as f:
        for idx,line in enumerate(f):
            if idx==count:
                break
            d=json.loads(line)
            d={col:d[col] for col in columns}
            data.append(d)
    data=pd.DataFrame(data)
    return data
data=readfile('arxiv-metadata-oai-2019.json',['id','abstract','categories','comments'])
data
```

```python
data['pages']=data['comments'].apply(lambda x:re.findall('[1-9][0-9]* pages',str(x)))
data=data[data['pages'].apply(len)>0]
data['pages']=data['pages'].apply(lambda x:int(x[0].replace(' pages','')))
data
```

```python
data['pages'].describe().astype(int)
```

```python
data['categories']=data['categories'].apply(lambda x:x.split(' ')[0])
data['categories']=data['categories'].apply(lambda x:x.split('.')[0])
plt.figure(figsize=(10,6))
data.groupby(['categories'])['pages'].mean().sort_values().plot(kind='barh')
plt.show()
```

```python
data_with_code=data[(data.comments.str.contains('github')==True)|(data.abstract.str.contains('github')==True)]
data_with_code['text']=data_with_code['abstract'].fillna('')+data_with_code['comments'].fillna('')
pattern='[a-zA-Z]+://github[^\s]*'
data_with_code['code_flag']=data_with_code['text'].str.findall(pattern).apply(len)
data_with_code
```

```python
data_with_code=data_with_code[data_with_code['code_flag']==1]
plt.figure(figsize=(12,6))
data_with_code.groupby('categories')['code_flag'].count().sort_values().plot(kind='barh')
plt.show()
```
